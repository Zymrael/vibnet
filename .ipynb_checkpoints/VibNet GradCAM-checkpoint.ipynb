## Heatmap

```python
i = iter(valloader)
x, y = next(i)
```

```python
y
```

```python
hook_output(m.conv_layers2[1])
```

```python
m.eval()
def hookDilatedHeads(m, x, y):
    with hook_output(m.conv_layers1[1]) as hook_a:
        with hook_output(m.conv_layers1[1], grad=True) as hook_a_grad:
            with hook_output(m.conv_layers2[1]) as hook_b:
                with hook_output(m.conv_layers2[1], grad=True) as hook_b_grad:
                    with hook_output(m.conv_layers3[1]) as hook_c:
                        with hook_output(m.conv_layers3[1], grad=True) as hook_c_grad:
                            preds = m(x)
                            preds[0,int(y)].backward()
    return hook_a, hook_b, hook_c, hook_a_grad, hook_b_grad, hook_c_grad
```

```python
ha,hb,hc,hga,hgb,hgc = hookDilatedHeads(m,x,y)
```

```python
ha,hb,hc = torch.mean(ha.stored[0],0).cpu(),torch.mean(hb.stored[0],0).cpu(),torch.mean(hc.stored[0],0).cpu()
```

```python
hga,hgb,hgc = torch.mean(hga.stored[0][0],0).cpu(),torch.mean(hgb.stored[0][0],0).cpu(),torch.mean(hgc.stored[0][0],0).cpu()
```

```python
def pad(arr,tsize=2048):
    n = (tsize-len(arr))//2 
    return np.pad(arr,n,'constant')
ha,hb,hc = pad(ha),pad(hb),pad(hc)
```

```python
hga,hgb,hgc = pad(hga),pad(hgb),pad(hgc)
```

```python
plt.figure(figsize=(15,5))
plt.plot(x[0].cpu(), color='black',zorder=1)
plt.bar(np.arange(2048),ha, width=0.8, color='red',zorder=2)
```

```python
plt.figure(figsize=(15,5))
plt.plot(x[0].cpu(), color='black',zorder=1)
plt.bar(np.arange(2048),hb, width=0.8, color='green',zorder=2)
```

```python
plt.figure(figsize=(15,5))
plt.plot(x[0].cpu(), color='black',zorder=1)
plt.bar(np.arange(2048),hc, width=0.8, color='blue',zorder=2)
```

```python
plt.figure(figsize=(15,5))
plt.bar(np.arange(2048),ha, width=0.8, color='red',zorder=1)
plt.bar(np.arange(2048),hb, width=0.8, color='green',zorder=2)
plt.bar(np.arange(2048),hc, width=0.8, color='blue',zorder=3)
plt.legend(['low dilation','medium dilation', 'high dilation'])
```
